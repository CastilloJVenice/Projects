```python
import scipy.stats as stats
import numpy as np
import math as math
from statsmodels.stats.proportion import proportions_ztest
```

"Five years ago,  3.9%  of children in a certain region lived with someone other than a parent. A sociologist wishes to test whether the current proportion is different. Perform the relevant test at the  5%  level of significance using the following data: in a random sample of  2,759  children,  119  lived with someone other than a parent."

There are two ways we can perform significance testing on proportions using python libraries:
First way is using the scipy library

```python
# state the hypothesis
# H0: p = .039 or 3.9%
# Ha: p != .039 or 3.9%

# find the given
x = 119     # no. of children who lived with someone other than a parent
n = 2759    # sample size
p = .039    # p0
alph = 0.05 # significance

# solve for the test statistic
p_hat = x/n
se = np.sqrt(p*(1-p)/n)

z = (p_hat-p)/se
print("se0 = "+"{:.4f}".format(se))
print("z = "+"{:.2f}".format(z))
```

We can find the significance using the Rejection Region or by finding the P-value

```python
# Rejection Region method

rr = alph/2              # since Ha: p != .039, we use two tailed test // for left tailed or right tailed test, no need to divide by 2
cz = stats.norm.ppf(rr)  # solve for the critical z-value
print("critical z = "+"{:.3f}".format(cz))
#cz_p = np.abs(cz)
#print("critical z = "+"{:.3f}".format(cz_p))
```

The sign can change depending on the test type (left tailed, right tailed, or two tailed). If it's two tailed then it's both positive and negative.
Use an absolute value function to make it positive if you really want to have the proper signs.

```python
# find if the test statistic is within the FTR

if(-1.96<z<1.96):
  print("Don't reject null hypothesis")
else:
  print("Reject null hypothesis")
```

```python
# p-value method
# since Ha: p != .039, we use two tailed test

pval = 1-stats.norm.cdf(z)      # Instead of .5 - 0.3686
pval = 2*pval                   # because two tailed test
print("p-value = "+"{:.4f}".format(pval))
```

```python
# find if p-value is greater than or lesser than significance

if(pval>=alph):
  print("Don't reject null hypothesis")
elif (pval<alph):
  print("Reject null hypothesis")
```

The second way is using the statsmodels library. More specifically the proportions_ztest() function

proportions_ztest() immediately solves for the test statistic and the p-value

parameters:
>  count = number of successes in a population (the x)
<br>  nobs = number of observations / the population (the n)
<br>  value = the null hypothesis
<br>  alternative = type of test ('two sided' is two tailed; 'smaller' if left-tailed; 'larger' if right-tailed )
<br>  prop_var = the value that the population proportion is tested against, input null hypothesis if one proportion test

```python
sm = proportions_ztest(count=x, nobs=n, value=p, alternative='two-sided', prop_var=p)
print(sm)
```

The values are saved as a list. sm[0] is the test statistic while sm[1] is the p-value

```python
print("z = "+"{:.2f}".format(sm[0]))
print("p-value = "+"{:.4f}".format(sm[1]))
```

```python
# find if p-value is greater than or lesser than significance

if(sm[1]>=alph):
  print("Don't reject null hypothesis")
elif (sm[1]<alph):
  print("Reject null hypothesis")
```
